```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, cv, bias, rbias, rct, acc = np.loadtxt('COLVAR', unpack=True, usecols=None)
```

```python
weights = np.exp(rbias)

fig = plt.figure(figsize=(15,4))
ax = fig.subplots(nrows=1, ncols=3)

ax[0].plot(time, cv)
ax[1].plot(time, bias)
ax[2].plot(time, weights)

plt.show()
```

```python
reduced_T=0.5
T=120.77*reduced_T # unit: K
kT = (6*1.38/1000)*T # mol/K

hist = np.histogram(cv, bins=30, weights=weights)
prob = hist[0].T/np.sum(hist[0].T)
free_energy = -kT*np.log(prob+1e-10)
mids = 0.5*(hist[1][1:]+hist[1][:-1])

fig = plt.figure(figsize=(5,4))
ax = fig.subplots(nrows=1, ncols=1)
ax.plot(mids, free_energy)
plt.show()
```
