```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time, coord, bias, rbias, rct, acc = np.loadtxt('COLVAR', unpack=True, usecols=None)
# time, coord = np.loadtxt('COLVAR', unpack=True, usecols=(0,1))
```

```python
plt.plot(time, coord)
plt.plot(time, 10*bias)
```

```python
weights = np.exp(rbias)
plt.plot(weights)
```

```python
reduced_T=0.5
T=120.77*reduced_T # unit: K
kT = (6*1.38/1000)*T # mol/K

hist = np.histogram(coord, bins=30, weights=weights)
prob = hist[0].T/np.sum(hist[0].T)
free_energy = -kT*np.log(prob+1e-10)
mids = 0.5*(hist[1][1:]+hist[1][:-1])
```

```python
plt.plot(mids, free_energy)
```
